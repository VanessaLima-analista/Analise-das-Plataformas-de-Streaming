**Convertendo coluna "duration"**

```python
df_disney_movies['duration'] = df_disney_movies['duration'].str.replace(' min', '').astype(float)
df_hulu_movies['duration'] = df_hulu_movies['duration'].str.replace(' min', '').astype(float)
```

```python
df_disney_shows['duration'] = df_disney_shows['duration'].str.replace(' Seasons', '')
df_disney_shows['duration'] = df_disney_shows['duration'].str.replace(' Season', '').astype(float)
df_hulu_shows['duration'] = df_hulu_shows['duration'].str.replace(' Seasons', '')
df_hulu_shows['duration'] = df_hulu_shows['duration'].str.replace(' Season', '').astype(float)
```

## 5. Juntando dataframes



```python
## Criando uma coluna chave para dfs

df_disney_movies['title_id'] = df_disney_movies['title'].str.lower().str.replace(' ','')
df_hulu_movies['title_id'] = df_hulu_movies['title'].str.lower().str.replace(' ','')
```

```python
df_movies_concat = pd.concat([df_disney_movies, df_hulu_movies])
```

```python
df_movies_concat
```

```python
df_movies = df_movies_concat.drop_duplicates(subset=['title_id'])
```

```python
df_movies
```

```python
df_movies.describe()
```

## 6. Medidas estatísticas

```python
Q1 = df_movies['duration'].quantile(0.25)
Q3 = df_movies['duration'].quantile(0.75)
IQR = Q3 - Q1
print(f'Q1: {Q1}\nQ3: {Q3}\nIQR: {IQR}')
```

```python
outliers_IQR = df_movies[(df_movies['duration'] < (Q1 - 1.5 * IQR)) | (df_movies['duration'] > (Q3 + 1.5 * IQR))]
outliers_IQR
```

```python
df_movies_tratado = df_movies[(df_movies['duration'] >= (Q1 - 1.5 * IQR)) & (df_movies['duration'] <= (Q3 + 1.5 * IQR))]
```

```python
df_movies_tratado.describe()
```
